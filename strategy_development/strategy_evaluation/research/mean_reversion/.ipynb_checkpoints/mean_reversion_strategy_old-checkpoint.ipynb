```python
import sys
sys.path.append('../data_exploration')
from stock_info import StockInfo
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define a function that returns the B&H returns versus the returns from my mean verversion strategy.
# Input parameters: [ ticker, period = [interaval, amount], scope, std deviation range]
# Outputs: [Buy and hold returns, Returns from strategy, risk adjusted returns]
```

```python
def mean_reversion(ticker, period, scope = 30):
    stock = StockInfo(ticker, start = period[0], end = period[1])

    # Buy and hold returns
    bah_investment = [ period[2] ]
    for i in range(scope, len(stock.data)):
        bah_perc_change = stock.data.iloc[i] / stock.data.iloc[i - 1]
        bah_investment.append(bah_investment[-1] * bah_perc_change.iloc[0])

    investment = period[2]
    strat_investment = [investment]
    position = False
    days_in_market = 0
    # Strategy Returns
    for i in range(scope, len(stock.data)):
        if position:
            days_in_market += 1
            perc_change = stock.data.iloc[i] / stock.data.iloc[i - 1]
            investment *= perc_change.iloc[0]
            strat_investment.append(investment)
            if stock.data.iloc[i].iloc[0] > stock.data.iloc[i - scope: i].mean().iloc[0] + 2 * np.sqrt(stock.data.iloc[i - scope : i].var().iloc[0]):
                position = False
        else:
            strat_investment.append(0)
        # If we do not have stock and the stock price is below 2 standard devations of the running mean over the past 'scope' time intervals:
        if not position and stock.data.iloc[i].iloc[0] < stock.data.iloc[i - scope: i].mean().iloc[0] - 2 * np.sqrt(stock.data.iloc[i - scope : i].var().iloc[0]):
            position = True

    # Work out time spent in the market
    perc_in_market = days_in_market / (len(stock.data) - scope)
    risk_adj_returns = ((investment / period[2] - 1) * 100) / perc_in_market
    

    plt.figure()
    plt.plot(strat_investment, label = 'Mean Reversion Stratgey Returns')
    plt.plot(bah_investment, label = 'Buy and Hold Returns', alpha = 0.4)
    plt.title(f"Mean Reversion Strategy versus Buy and Hold Strategy for {ticker}")
    plt.xlabel('Time')
    plt.ylabel('Investment Value')
    plt.legend()
    plt.show()
            
    print('Buy and Hold Returns:',round(bah_investment[-1], 2), f"  percentage_increase:  {(bah_investment[-1] / period[2] - 1) * 100:.2f} %" ,'\nStrategy Returns:',round(investment, 2),f"      percentage_increase:  {(investment / period[2] - 1) * 100:.2f} %",f" Risk-Adjusted Returns: {risk_adj_returns:.2f} %")
    return [round(bah_investment[-1], 2), round(investment, 2)]
            
    
```

```python
period = ['2007-01-01', '2020-01-01', 100]
```

```python
mean_reversion('AAPL', period)
```

```python
periods = [
    ['2010-01-02', '2011-01-01', 100],
    ['2011-01-02', '2012-01-01', 100],
    ['2012-01-02', '2013-01-01', 100],
    ['2013-01-02', '2014-01-01', 100],
    ['2014-01-02', '2015-01-01', 100],
    ['2015-01-02', '2016-01-01', 100],
]
```

```python
for period in periods:
    mean_reversion('AAPL', period)
```
