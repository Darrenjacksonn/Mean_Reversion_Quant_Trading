```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import sys
sys.path.append('../')
from strategy_formulation.strategy.tilted_mean_reversion_strategy import MeanReversion
from strategy_formulation.research.stock_info.stock_info import sap500_tickers
import numpy as np
import yfinance as yf
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Four 10 year blocks
# The below parameters do not change
```

```python
periods = [ 
    ['1984-01-02', '1994-01-01', 100], 
    ['1994-01-02', '2004-01-01', 100], 
    ['2004-01-02', '2014-01-01', 100], 
    ['2014-01-02', '2024-01-01', 100],
]
index_labels = [f"{start} to {end}" for start, end, _ in periods]
```

```python
# The below code is used to pull tickers for analysis. By default we use the top 50 of the S&P500 for analysis
```

```python
tickers = sap500_tickers(0,50)
dict_of_tickers = dict()
for ticker in tickers:
    dict_of_tickers[ticker] = MeanReversion(ticker)
burner_set = set()
for key in dict_of_tickers:
    if dict_of_tickers[key].data.empty:
        burner_set.add(key)
for el in burner_set:
    del dict_of_tickers[el]
dict_of_tickers
```

```python
relevant_metrics = ['B&H % increase','Strat risk-adj % increase', 'No. of trades', 'Win / Loss Ratio']
output_data = {(key, metric) : [] for key in dict_of_tickers for metric in relevant_metrics}
output_data
```

```python
for key in dict_of_tickers:
    for period in periods:
        result = dict_of_tickers[key].tilted_mean_reversion(period, scope = 30, buy_range = 2, sell_range = 2, graph = False, analysis = False)
        for metric in relevant_metrics:
            output_data[(key, metric)].append(result[metric])
output_df = pd.DataFrame(output_data)
output_df.index = index_labels
output_df
```
