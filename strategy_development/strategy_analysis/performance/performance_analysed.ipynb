```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_df = pd.read_csv('report1_performance_s30_b2_s2.csv', header = [0,1], index_col = 0)
my_df
```

```python
my_df['MSFT']
```

```python
risk_adj_gains = my_df.xs(key = 'Strat risk-adj % increase', level = 1, axis = 1)
mean_risk_adj_gains = risk_adj_gains.mean().to_frame()
mean_risk_adj_gains
```

```python
bah_gains = my_df.xs(key = 'B&H % increase', level = 1, axis = 1)
mean_bah_gains = bah_gains.mean().to_frame()
mean_bah_gains
```

```python
comparison = mean_risk_adj_gains - mean_bah_gains
comparison
```
