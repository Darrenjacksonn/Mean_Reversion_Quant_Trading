```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import sys
sys.path.append('../')
from performance.performance_function import performance_analysis
sys.path.append('../../')
from strategy_formulation.strategy.tilted_mean_reversion_strategy import MeanReversion
from strategy_formulation.research.stock_info.stock_info import sap500_tickers
import matplotlib.pyplot as plt
import numpy as np
```

```python
tickers_range = [0,10]

relevant_metrics = ['B&H % increase','Strat risk-adj % increase', 'No. of trades', 'Win / Loss Ratio']
```

```python
# Tweak scope, and plot scope versus risk adj returns
```

```python
period = ['2010-01-01', '2015-01-01', 100]
tickers = sap500_tickers(tickers_range[0],tickers_range[1])
scope_range = [20,252]

plt.figure(figsize=(12,8))
plt.xlabel('Scope')
plt.ylabel('Risk Adj net Returns')
plt.title('Scope versus Returns')

for ticker in tickers:
    ticker_name = MeanReversion(ticker)
    plotted_scope = np.arange(scope_range[0], scope_range[1])
    plotted_risk_adj_returns = []
    for scope in range(scope_range[0], scope_range[1]):
        result = ticker_name.tilted_mean_reversion(period, scope = scope, graph = False, analysis = False)
        plotted_risk_adj_returns.append(result['Strat risk-adj % increase']  - result['B&H % increase'])
    plt.plot(plotted_scope, plotted_risk_adj_returns, label = ticker)
plt.legend()
plt.show()
```
