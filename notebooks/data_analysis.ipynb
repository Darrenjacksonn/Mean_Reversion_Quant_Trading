```python
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
SPL = yf.download('SPY')
SPL
```

```python
SPL = yf.download('SPY', interval = '1d')
SPL
```

```python
SPL['Close'].to_frame().plot()
```

```python
SPL['Adj Close'].to_frame().plot()
```

```python
spl_close = SPL['Close'].to_frame()
spl_close
```

```python
spl_daily_returns = SPL['Close'].pct_change().mul(100).to_frame().dropna()
spl_daily_returns = spl_daily_returns.rename(columns = {'Close' : 'Daily % Returns'})
spl_daily_returns
```

```python
spl_daily_returns.plot(kind = 'hist', bins = 1000, title = 'Daily Percentage Returns')
plt.show()
```

```python
average_daily_returns = spl_daily_returns.mean()
average_daily_returns
```

```python
# If you invested â‚¬1 in 1993, you would have: 7812
returns = 1 * average_daily_returns * 7812
returns
```
